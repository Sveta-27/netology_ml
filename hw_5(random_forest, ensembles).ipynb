# Modeling

### Random forest

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state = 1)
```

```python
rf = RandomForestRegressor(n_estimators=2000, max_depth=30, random_state = 30)
rf.fit(df_train[df_train.columns.difference(['SalePrice'])], df_train['SalePrice'])   
```

```python
y_pred_train = rf.predict(df_train[df_train.columns.difference(['SalePrice'])])
y_pred_test = rf.predict(df_test[df_test.columns.difference(['SalePrice'])])

print(r2_score(y_pred_train, df_train['SalePrice']))
print(r2_score(y_pred_test, df_test['SalePrice']))
```

```python
# Важности признаков случайного леса
imp = pd.Series(rf.feature_importances_)
imp.sort_values(ascending=False)
```

### Staking

```python
X_train = df_train[df_train.columns.difference(['SalePrice'])]
# X_train.head()
```

```python
X_test = df_test[df_test.columns.difference(['SalePrice'])]
# X_test.head()
```

```python
y_train = df_train['SalePrice']
# y_train.head()
```

```python
y_test = df_test['SalePrice']
# y_test.head()
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
r = StackingRegressor(
    [
        ('lr1', LinearRegression()),
        ('rf', RandomForestRegressor(n_estimators=200, max_depth=30, random_state = 10)),
        ('lr2', LinearRegression())
    ],
LinearRegression())

print(r.fit(X_train, y_train).score(X_train, y_train))
print(r.fit(X_train, y_train).score(X_test, y_test))
```
